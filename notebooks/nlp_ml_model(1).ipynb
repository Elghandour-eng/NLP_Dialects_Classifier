```python
result_df=pd.read_csv("/kaggle/input/nlp-ml123/data.csv")
```

```python
le = LabelEncoder()
result_df['dialect'] = le.fit_transform(result_df['dialect'])
```

```python
result_df['text'] = result_df['text'].fillna('')
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        result_df['text'], result_df['dialect'], random_state=42, test_size=0.1, shuffle=True)
```

```python
tfidf = TfidfVectorizer(
    max_df=0.95, 
    ngram_range=(1, 5),
    max_features= 25000,
    analyzer='word')
```

```python
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

LogisticRegression

```python
lr = LogisticRegression(multi_class='ovr')
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
accuracy =accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Model accuracy:", accuracy)
print("Classification report:\n",report)
```

RandomForestClassifier

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Modelaccuracy:", accuracy)
print("Classification report:\n", report)
```

lgb

```python
lgb_clf = lgb.LGBMClassifier()

lgb_clf.fit(X_train, y_train)
```

```python
y_pred = lgb_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Modelaccuracy:", accuracy)
print("Classification report:\n", report)
```
